```python
#imports of numpy and plotting.
import numpy as np
import matplotlib.pyplot as plt

#Plot figure style
plt.style.use('seaborn-dark-palette')

#Changing the default size of the figure.
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
mu, sigma = 0,0.1 #mean and standard deviation.
s = np.random.normal(mu,sigma, 1000)
```

```python
#Checking the length of s.
len(s)
```

```python
#cutting s down to a size that is readable for the user.
#the ':' will give you the numbers from index 0 to 49(50)
s[:50]
```

```python
abs(mu - np.mean(s))
```

```python
abs(sigma - np.std(s,ddof=1))
```

```python
count, bins, ignored = plt.hist(s, 30, density = True)# hist is short for histogram
x = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(bins - mu)**2/(2 * sigma**2))#here I am following the pep8 docs
plt.plot(bins,x,linewidth=2, color='y')
plt.show()
```
